```python
from tkinter import *


class Project(Frame):
    def __init__(self, root):
        super(Project, self).__init__(root)
        self.build()

    def build(self):
        self.formula = "0"
        self.lbl = Label(text=self.formula, font=("Times New Roman", 21, "bold"), bg="#808080", foreground="#FFF")
        self.lbl.place(x=11, y=50)

        knop = [
            "C", "DEL", "X^2", "/",
            "7", "8", "9", "*",
            "4", "5", "6", "+",
            "1", "2", "3", "-",
            "%", "0", ".", "="
        ]

        x = 10
        y = 140
        for bat in knop:
            cf = lambda x=bat: self.logica(x)
            Button(text=bat, bg="#FFF", font=("Helvetica", 15), command=cf).place(x=x, y=y, width=115, height=79)
            x += 117
            if x > 400:
                x = 10
                y += 81

    def logica(self, operation):
        if operation == "C":
            self.formula = ""
        elif operation == "DEL":
            self.formula = self.formula[0:-1]
        elif operation == "X^2":
            if self.formula[-1] == '+' or self.formula[-1] == '-' or self.formula[-1] == '*' or self.formula[-1] == '/': pass
            else: self.formula = str((eval(self.formula))**2)
        elif operation == "=":
            if self.formula[-1] == '+' or self.formula[-1] == '-' or self.formula[-1] == '*' or self.formula[-1] == '/': pass
            else: self.formula = str(eval(self.formula))
        elif operation == "+" or operation == "-" or operation == "/" or operation == "*":
            if self.formula[-1] == '+' or self.formula[-1] == '-' or self.formula[-1] == '*' or self.formula[-1] == '/':
                self.formula = self.formula[0:-1]
            self.formula += operation
        else:
            if self.formula == "0":
                self.formula = ""
            self.formula += operation
        self.update()

    def update(self):
        if self.formula == "":
            self.formula = "0"
        self.lbl.configure(text=self.formula)


if __name__ == '__main__':
    cal = Tk()
    cal["bg"] = "#808080"
    cal.geometry("485x540+600+200")
    cal.title("Калькулятор")
    cal.resizable(False, False)
    app = Project(cal)
    app.pack()
    cal.mainloop()
```

```python
import tkinter as tk

class MyApp(tk.Frame):
    def __init__(self, master=None, **options):
        super().__init__(master, **options)
        self.build()

    def build(self):
        # Устанавливаем отступы между столбцами в 5 пикселей
        self.grid_columnconfigure(0, pad=5)
        self.grid_columnconfigure(1, pad=5)
        self.grid_columnconfigure(2, pad=5)

        # Создаем виджеты
        tk.Label(self, text='Column 1').grid(row=0, column=0)
        tk.Label(self, text='Column 2').grid(row=0, column=1)
        tk.Label(self, text='Column 3').grid(row=0, column=2)

root = tk.Tk()
app = MyApp(root)
app.grid()
root.mainloop()
```

```python
import tkinter as tk

class MyApp(tk.Frame):
    def __init__(self, master=None, **options):
        super().__init__(master, **options)
        self.entry1 = None
        self.entry2 = None
        self.button = None
        self.label = None
        self.build()

    def build(self):
        self.columnconfigure(0, weight=1)

        # Создаем поля ввода
        self.entry1 = tk.Entry(self)
        self.entry1.grid(row=0, column=0)

        self.entry2 = tk.Entry(self)
        self.entry2.grid(row=1, column=0)

        # Создаем кнопку
        self.button = tk.Button(self, text='Выполнить', command=self.button_click)
        self.button.grid(row=2, column=0)

        # Создаем метку
        self.label = tk.Label(self, text='')
        self.label.grid(row=3, column=0)

    def button_click(self):
        # Получаем значения из полей ввода
        value1 = self.entry1.get()
        value2 = self.entry2.get()

        # Вызываем функцию с передачей значений
        result = self.my_function(value1, value2)

        # Обновляем метку
        self.label.config(text=result)

    def my_function(self, value1, value2):
        # Это просто пример функции, которая складывает два числа
        result = int(value1) + int(value2)
        return f'Результат: {result}'

root = tk.Tk()
app = MyApp(root)
app.pack(expand=True, fill='both')
root.mainloop()
```
